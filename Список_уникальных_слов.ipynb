```python
with open('salinger1.txt', encoding='utf-8') as f:
  f_text = f.read().lower()
```

```python
f_text
```

Убираем лишнюю пунктуацию

```python
f_text = f_text.replace(",",'')
```

```python
f_text = f_text.replace('.', '')
```

```python
words_raw = f_text.split()
```

```python
words_raw
```

```python
len(words_raw)
```

```python
words = []

for word in words_raw:
  if word not in words:
    words.append(word)
```

```python
words.sort() #сортируем по алфавиту
```

```python
words
```

```python
len(words)
```
